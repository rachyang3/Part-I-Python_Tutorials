# 加载matplotlib库

我们需要绘制图表的大部分功能都在matplotlib.pyplot中。 习惯上将其导入为 **plt**

````python
import matplotlib.pyplot as plt
import numpy as np
````

# Plotting

````python
x = [0,1,2]
y = [0,1,4]
````

````python
fig = plt.figure()           #定义图表
axes = fig.add_subplot(111)  #定义子图，参数是1行1列第1个绘图

axes.plot(x, y)  #x横坐标，y纵坐标

plt.show()
````

#### Matplotlib 里的常用类的包含关系为 Figure -> Axes -> (Line2D, Text, etc.)。一个Figure对象可以包含多个子图(Axes)，在matplotlib中用Axes对象表示一个绘图区域，可以理解为子图。
#### 可以使用subplot()快速绘制包含多个子图的图表，它的调用形式如下：
#### subplot(numRows, numCols, plotNum)  参数是几行几列第几个图

````python
plt.figure(1) # 创建图表1
plt.figure(2) # 创建图表2
ax1 = plt.subplot(211) # 子图1为2行1列的第1个图
ax2 = plt.subplot(212) # 子图2为2行1列的第2个图

x = np.linspace(0, 3, 100)
for i in range(5):

    plt.figure(1)  #❶ # 选择图表1
    plt.plot(x, np.exp(i*x/3))   #没有设置子图，用默认值

    plt.sca(ax1)   #❷ # 选择图表2的子图1
    ax1.set_title('figure2,ax1')
    plt.plot(x, np.sin(i*x))

    plt.sca(ax2)  # 选择图表2的子图2
    plt.plot(x, np.cos(i*x))
plt.show()
````

#操作理解subplot

````python
plt.subplot(221) # 第一行的左图
plt.subplot(222) # 第一行的右图
plt.subplot(212) # 第二整行
plt.show()
````

操作理解subplot2grid

````python
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2, 0))
ax5 = plt.subplot2grid((3,3), (2, 1))
plt.show()
````

## 控制 Line Style

````python
fig = plt.figure()
axes = fig.add_subplot(111)
x_highres=np.arange(1,20)
y_highres=x_highres**2
axes.plot(x_highres, y_highres, "r--")
#axes.plot(x_highres, y_highres, color="red", linestyle='dashed')

plt.show()
````

````python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres, color="red", linestyle='dashed',  linewidth=3)

plt.show()
````

````python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres,color="red", linestyle='dashed',  linewidth=3, marker='o',
         markerfacecolor='blue', markersize=5)

plt.show()
````

## 标题和网格（Title and Grid）

````python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres,color="red", linestyle='dashed',  linewidth=3, marker='o',
         markerfacecolor='blue', markersize=5)

axes.set_title('$y=x^2$') ## Notice you can use LaTeX Code
                          ## for more about LaTeX check the Tutorial about Markdown and LaTeX
axes.grid()

plt.show()
````

# 坐标轴标签（Axis Labels）

````python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres,color="red", linestyle='dashed',  linewidth=3, marker='o',
         markerfacecolor='blue', markersize=5)

axes.set_title('$y=x^2$')
axes.grid()
axes.set_xlabel('长度')    #设置x坐标标签
axes.set_ylabel('y')    #设置y坐标标签

plt.show()
````

## 控制图大小（ Figure Size）

````python
fig = plt.figure(figsize=(12,8))  #图的大小（in inches）
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres,color="red", linestyle='dashed',  linewidth=3, marker='o',
         markerfacecolor='blue', markersize=5)

axes.set_title('$y=x^2$')
axes.grid()
axes.set_xlabel('x')
axes.set_ylabel('y')

plt.show()
````
